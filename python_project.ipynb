```python
import seaborn as sns
import pandas as pd
mpg = sns.load_dataset("mpg")
sns.set_theme()
```

●	**Create a table summarizing fuel economy by country of origin.**

    The visual shows that cars from Japan has the highest mpg, while American cars has the lowest mpg.

●	**Create a visual showing how fuel economy has changed over time.**

    The visual shows that the newer car models usually has higher mpg than the older models.

*● Create a visual showing the relationship between horsepower and fuel economy.*

```python
sns.lineplot(data=mpg,x="horsepower",y="mpg")
```

*● Create a table summarizing fuel economy by country of origin.*

```python
sns.catplot(data=mpg,x="origin",y="mpg",kind="bar",errorbar=None)
```

*● Create a visual showing how fuel economy has changed over time.*

```python
sns.lineplot(data=mpg,x="model_year",y="mpg")
```

*● Create two other visuals or tables related to fuel economy.*

● Top 5 cars with highest mpg.

```python
top5 = mpg.groupby("name").mpg.sum().sort_values(ascending=False).head(5)
```

```python
top5x = top5.to_frame().reset_index()
```

```python
sns.catplot(data=top5x,x="mpg",y="name",kind="bar",errorbar=None)
```

● Mpg per pound compared with total weight.

```python
mpg['mpgperpound'] = mpgperp = mpg.mpg / mpg.weight
```

```python
sns.lineplot(data=mpg,x="weight",y="mpgperpound")
```
